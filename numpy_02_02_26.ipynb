```python
a=np.array([[1,2,3],[3,2,1]])
b=np.array([[4,5],[5,4],[4,5]])
```

```python
#a*b
mul1=np.matmul(a,b)
mul1
```

```python
print(mul1.ndim)
print(mul1.shape)
print(mul1.size)
print(mul1.dtype)
print(mul1.itemsize)
print(mul1.nbytes)
print(type(mul1))
```
