#reshape

```python
dd=np.arange(10).reshape(5,2)
print(dd)
```

```python
print(type(dd))
print(dd.ndim)
print(dd.dtype)
print(dd.nbytes)
print(dd.shape[0])
```

#Slicing

```python
ee=np.arange(100)
```

```python
ff=ee[10:20]
print(ff)
```

```python
ff[0]=100
print(ff)
```

```python
print(ee)
```

```python
a=ee[10:20].copy()
print(a)
```

```python
a[0]=10
print(a)
print(ee)
```

```python
ff[0]=10
ee
```

```python
ee[::5]
```

```python
ee[::-5]
```

```python
ee[10]=100
print(ee)
```

```python
idx=np.argwhere(ee==100)[0][0]
print(idx)
```

```python
ee[idx]=10
ee
```
