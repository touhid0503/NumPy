```python

a = np.array([10, 20, 30, 40])

# Normal indexing
x = a[1]
print("a[1]:", x)
print("type:", type(x))
print("shape:", x.shape)
print(x.ndim)
print(x.dtype)
print(x.nbytes)
print(x.size)

# Fancy indexing
y = a[[0,2,3]]
print("\na[[1]]:", y)
print("type:", type(y))
print("shape:", y.shape)
print(y.ndim)
print(y.dtype)
print(y.nbytes)
print(y.shape[0])
print(len(y))
print(y.size)
```

```python
a[[True,True,False,True]]
```

```python
a[a<30]
```

```python
a[(a > 10) & (a < 40)]
```

#copy vs view

```python
a=np.array([1,2,3,4])
b=a.copy()
b[0]=10
print(a)
print(b)
```

```python
b[[0,2,3]]=15
print(b)
print(a)
```

#Seed : ufuncs

```python
B=np.random.rand(1000000)
%timeit B.sum()
%timeit np.sum(B)
```

```python
def mysum(g):
  s=0
  for x in g:
      s+=x
  return s
```

```python
%timeit mysum(B)
```
